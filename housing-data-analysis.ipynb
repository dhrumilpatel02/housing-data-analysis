```python
#Load libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('./HousingNew.csv')
df.head()
```

```python
df.describe()
```

```python
plt.plot(df.price, label='price')

#Titles
plt.title('Price')
plt.xlabel('ID')
plt.ylabel(' Selling Price of the House')

#Legend 
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()

plt.plot(df.lotsize, label='lotsize')

#Titles
plt.title('Lotsize')
plt.xlabel('ID')
plt.ylabel('measured in sq meters')

#Legend 
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```
